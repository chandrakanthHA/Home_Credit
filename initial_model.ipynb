```python
data = pd.read_csv('Home_Loan/application_train.csv.tar.gz')
```

```python
# Function to calculate missing values in each column
def missing_values(df):
    # Total missing values
    mis_values = df.isnull().sum()
    # percentage of total missing values
    mis_values_percent = (100 * mis_values) / len(df)
    # table with missing values
    mis_values_table = pd.concat([mis_values, mis_values_percent], axis=1)
    
    # rename the columns
    mis_values_table = mis_values_table.rename(columns={0 : 'Missing Values', 1 : '% of Missing Values'})
    
    # sorting the table per percentage
    mis_values_table = mis_values_table[mis_values_table.iloc[:,1] != 0].sort_values(
        '% of Missing Values', ascending = False).round(2)
    
    return mis_values_table
```

```python
missing_values_table = missing_values(train_data)
```

```python
train_data_clean = train_data.dropna()
```

```python
train_data_clean.shape
```

```python
missing_values(train_data_clean)
```

```python
model_data = train_data_clean[['TARGET', 'NAME_CONTRACT_TYPE','CODE_GENDER', 'FLAG_OWN_CAR','FLAG_OWN_REALTY',
                             'CNT_CHILDREN', 'AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'AMT_GOODS_PRICE',
                             'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'DAYS_BIRTH',
                             'DAYS_EMPLOYED', 'REGION_RATING_CLIENT_W_CITY', 'EXT_SOURCE_2', 'FLAG_DOCUMENT_2', 'FLAG_DOCUMENT_21']]
```

```python
model_data.head()
```

```python
cat_var = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 
                     'NAME_INCOME_TYPE','NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS']
```

```python
final_model_data = pd.get_dummies(model_data, columns=cat_var, drop_first=True)
```

```python
final_model_data.head()
```

```python
final_model_data.shape
```

```python
# splitting the data
from sklearn.model_selection import train_test_split

X = final_model_data.drop('TARGET', axis =1)
y = final_model_data['TARGET']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state = 50)
```

```python
# importing Regression models and evaluation matrices
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Logisticregression Model
logreg = LogisticRegression(class_weight='balanced', random_state=50)
logreg.fit(X_train, y_train)
```

```python
y_pred_log = logreg.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_log))
```

```python
accuracy_score(y_test, y_pred_log)
```

```python
print(confusion_matrix(y_test, y_pred_log))
```

```python
# KNN model
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean', weights='distance')
knn.fit(X_train, y_train)
```

```python
y_pred_knn = knn.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_knn))
```

```python
print(confusion_matrix(y_test, y_pred_knn))
```

```python
# Random forest model
# Create the model with 500 trees
R_forest = RandomForestClassifier(n_estimators=500, 
                               random_state=50, 
                               max_features = 'sqrt',  
                               n_jobs=-1, verbose = 1)
```

```python
R_forest.fit(X_train, y_train)
```

```python
n_nodes = []
max_depths = []

# Stats about the trees in random forest
for ind_tree in R_forest.estimators_:
    n_nodes.append(ind_tree.tree_.node_count)
    max_depths.append(ind_tree.tree_.max_depth)
    
print(f'Average number of nodes {int(np.mean(n_nodes))}')
print(f'Average maximum depth {int(np.mean(max_depths))}')
```

```python
# Testing predictions (to determine performance)
y_pred_forest = R_forest.predict(X_test)
y_probs_forest = R_forest.predict_proba(X_test)[:, 1]
```

```python
print(y_pred_forest)
print(y_probs_forest.mean())
```

```python
accuracy_score(y_test, y_pred_forest)
```

```python
print(classification_report(y_test, y_pred_forest))
```

```python
print(confusion_matrix(y_test, y_pred_forest))
```
